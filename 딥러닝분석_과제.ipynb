# Loading Data

* date: Date house was sold #팔린 날짜
* price: Price of the sold house # 팔린 가격
* bedrooms: Number of Bedrooms # 침실의 수
* bathrooms: Number of bathrooms # 화장실 수
* sqft_living: Square footage of the living space # 실거주 면적 평방 피트
* sqrt_lot: Square footage of the lot # 평방 피트
* floors: Total floors in the house #층수
* waterfront: Whether the house is on a waterfront(1: yes, 0: no) # 물 뷰인가
* view: special view? # 스페셜 뷰인가?
* condition: Condition of the house 
* sqft_above: Square footage of house apart from basement
* sqft_basement: Square footage of the basement
* yr_built: Built year
* yr_renovated: Year when the house was renovated 
* street: street of the house
* city: city of the house
* statezip: zipcode of the house
* country: country of the house
